````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
def load_csv_file(file_path, sep, encoding='utf-8'):
    """
    Esta fun√ß√£o carrega um arquivo CSV, realiza o tratamento da coluna 'SEMANA' e retorna um DataFrame.

    Argumentos:
        file_path (str): Caminho para o arquivo CSV.
        sep (str): Separador de campo no CSV.

    Retorno:
        pandas.DataFrame: DataFrame com as colunas do CSV, incluindo a coluna 'SEMANA' tratada.

    Exce√ß√µes:
        FileNotFoundError: Se o arquivo CSV n√£o for encontrado.
        Exception: Se ocorrerem outros erros durante a execu√ß√£o da fun√ß√£o.
    """
    try:
        df = pd.read_csv(file_path, sep=sep, encoding=encoding)

        if "SEMANA" in df.columns:
            try:
                df["SEMANA"] = pd.to_datetime(df["SEMANA"])
            except Exception as e:
                print(f"Erro ao converter a coluna 'SEMANA' para datetime: {e}")

        return df
    except FileNotFoundError as e:
        print(f"Arquivo CSV n√£o encontrado: {file_path}")
        raise
    except Exception as e:
        print(f"Erro inesperado ao carregar o CSV: {e}")
        raise
````

````python
def rename_columns(df):
    """
      Renomeia as colunas de um DataFrame Pandas de acordo com um dicion√°rio predefinido de novos nomes.

      Este fun√ß√£o verifica primeiro se as colunas especificadas existem no DataFrame. Se existirem,
      as colunas s√£o renomeadas de acordo com as regras definidas no dicion√°rio 'colunas_novas'.
      Caso contr√°rio, uma mensagem informando que nenhuma coluna foi encontrada para renomear √© exibida.

      Par√¢metros:
      df (pandas.DataFrame): O DataFrame cujas colunas devem ser renomeadas.

      Retorna:
      pandas.DataFrame: O DataFrame com as colunas renomeadas, se aplic√°vel.

    """

    # Dicion√°rio das novas colunas com seus nomes
    colunas_novas = {
        "SS": "sessoes_na_loja",
        "SST": "sessoes_na_vertical",
        "VPD": "sessoes_de_visualizacao_de_produto",
        "ATC": "sessoes_de_adicao_de_produto_ao_carrinho",
        "PTC": "sessoes_de_proceder_ao_checkout",
        "OP": "sessoes_de_pedidos_solicitados",
        "OPC": "sessoes_de_pedidos_confirmados"
    }

    # Lista de colunas que realmente existem no DataFrame
    colunas_existentes = [coluna for coluna in colunas_novas.keys() if coluna in df.columns]

    if colunas_existentes:
        # Renomeando as colunas existentes
        df.rename(columns={coluna: colunas_novas[coluna] for coluna in colunas_existentes}, inplace=True)
        #print("Colunas renomeadas com sucesso: ", colunas_existentes)

    return df
````

### Gr√°fico Geral: Sess√µes Semanais com Linha de M√©dia na Vertical Restaurantes

- **T√≠tulo:** Sess√µes Semanais em Restaurantes
- **Descri√ß√£o:** Este gr√°fico mostra o total de sess√µes (SST) em restaurantes a cada semana. As barras representam a quantidade de sess√µes registradas semanalmente, com uma linha vermelha pontilhada indicando a m√©dia semanal. Os valores num√©ricos em cima das barras facilitam a visualiza√ß√£o dos dados espec√≠ficos.

````python
df_data_weekly = df_visao_geral_restaurantes_sessoes

# Convertendo colunas
df_data_weekly['SST'] = df_data_weekly['SST'].astype('int64')
df_data_weekly['SS'] = df_data_weekly['SS'].astype('int64')
df_data_weekly['OP'] = df_data_weekly['OP'].astype('int64')
df_data_weekly['OPC'] = df_data_weekly['OPC'].astype('int64')

numeric_cols = df_data_weekly.select_dtypes(include='number').columns
weekly_data = df_data_weekly.groupby('SEMANA')[numeric_cols].sum().reset_index()

sns.set()

# C√°lculo da m√©dia semanal
weekly_mean = weekly_data['SST'].mean()

def format_number(value):
    return f'{value:,.0f}'.replace(',', 'X').replace('.', ',').replace('X', '.')

plt.figure(figsize=(14, 7))
sns.barplot(x='SEMANA', y='SST', data=weekly_data)
plt.axhline(weekly_mean, color='red', linestyle='--', label=f'M√©dia: {format_number(weekly_mean)}')
plt.title('Sess√µes Semanais na Vertical Restaurantes', fontsize=16)
plt.xlabel('Semana', fontsize=14)
plt.ylabel('Total de Sess√µes na Vertical (SST)', fontsize=14)
plt.xticks(rotation=45)
for index, row in weekly_data.iterrows():
    plt.text(index, row.SST, format_number(row.SST), color='black', ha="center", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# M√©dia e Mediana da Vis√£o Geral Restaurantes: Sessoes
# Calcular a m√©dia e a mediana para cada coluna num√©rica
mean_values = df_visao_geral_restaurantes_sessoes.mean(numeric_only=True).round(0)
median_values = df_visao_geral_restaurantes_sessoes.median(numeric_only=True).round(0)

stats_df_visao_geral_restaurantes_sessoes = pd.DataFrame({
    'Estat√≠stica': ['M√©dia', 'Mediana'],
    'SST': [mean_values['SST'].astype('int64'), median_values['SST'].astype('int64')],
    'SS': [mean_values['SS'].astype('int64'), median_values['SS'].astype('int64')],
    'VPD': [mean_values['VPD'].astype('int64'), median_values['VPD'].astype('int64')],
    'ATC': [mean_values['ATC'].astype('int64'), median_values['ATC'].astype('int64')],
    'PTC': [mean_values['PTC'].astype('int64'), median_values['PTC'].astype('int64')],
    'OP': [mean_values['OP'].astype('int64'), median_values['OP'].astype('int64')],
    'OPC': [mean_values['OPC'].astype('int64'), median_values['OPC'].astype('int64')]
})

stats_df_visao_geral_restaurantes_sessoes.head()
````

### Gr√°fico Geral: Sess√µes Mensais com Linha de M√©dia na Vertical Restaurantes

- **T√≠tulo:** Sess√µes Mensais em Restaurantes
- **Descri√ß√£o:** Este gr√°fico mostra o total de sess√µes (SST) em restaurantes a cada m√™s. As barras representam a quantidade de sess√µes registradas mensalmente, com uma linha vermelha pontilhada indicando a m√©dia mensal. Os valores num√©ricos em cima das barras facilitam a visualiza√ß√£o dos dados espec√≠ficos.

````python
df_data_month_sst = df_visao_geral_restaurantes_sessoes

df_data_month_sst['SST'] = df_data_month_sst['SST'].astype('int64')
df_data_month_sst['SS'] = df_data_month_sst['SS'].astype('int64')
df_data_month_sst['OP'] = df_data_month_sst['OP'].astype('int64')
df_data_month_sst['OPC'] = df_data_month_sst['OPC'].astype('int64')

df_visao_geral_restaurantes_sessoes['MES'] = df_visao_geral_restaurantes_sessoes['SEMANA'].dt.to_period('M')

numeric_cols = df_visao_geral_restaurantes_sessoes.select_dtypes(include='number').columns

monthly_data = df_visao_geral_restaurantes_sessoes.groupby('MES')[numeric_cols].sum().reset_index()

meses_para_projecao = monthly_data[monthly_data['MES'].astype(str).isin(['2024-04', '2024-05'])]

if not meses_para_projecao.empty:
    projection_june = meses_para_projecao['SST'].mean()

    projection_data = pd.DataFrame({'MES': ['2024-06 (Proje√ß√£o)'], 'SST': [int(projection_june)]})

    monthly_data = pd.concat([monthly_data, projection_data], ignore_index=True)

    sns.set()

    monthly_mean = monthly_data[monthly_data['MES'] != '2024-06 (Proje√ß√£o)']['SST'].mean()

    def format_number(value):
        return f'{value:,.0f}'.replace(',', 'X').replace('.', ',').replace('X', '.')

    monthly_data['color'] = ['Proje√ß√£o' if x == '2024-06 (Proje√ß√£o)' else 'Real' for x in monthly_data['MES']]

    # Calcular a diferen√ßa percentual entre os meses
    monthly_data['percent_change'] = monthly_data['SST'].pct_change() * 100
    monthly_data['percent_change'] = monthly_data['percent_change'].round(2).astype(str) + '%'

    # Remover a diferen√ßa percentual da proje√ß√£o
    monthly_data.loc[monthly_data['MES'] == '2024-06 (Proje√ß√£o)', 'percent_change'] = ''

    plt.figure(figsize=(14, 7))
    sns.barplot(x='MES', y='SST', data=monthly_data, hue='color', dodge=False, palette={'Proje√ß√£o': 'grey', 'Real': '#4c72b0'})
    plt.axhline(monthly_mean, color='red', linestyle='--', label=f'M√©dia: {format_number(monthly_mean)}')
    plt.title('Sess√µes Mensais em Restaurantes', fontsize=16)
    plt.xlabel('M√™s', fontsize=14)
    plt.ylabel('Total de Sess√µes na Vertical (SST)', fontsize=14)
    plt.xticks(rotation=45)
    for index, row in monthly_data.iterrows():
        plt.text(index, row.SST, format_number(row.SST), color='black', ha="center", fontsize=12)
        if row['percent_change'] != '':
            plt.text(index, row.SST * 0.92, row['percent_change'], color='white', ha="center", fontsize=10)
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("Os meses 2024-04 e 2024-05 n√£o est√£o presentes nos dados para calcular a proje√ß√£o.")
````

# **Caso aumentarmos a dispobilidade dos produtos, quanto projetivo de convers√£o teremos?**

Para projetar o aumento na convers√£o com base na melhoria da disponibilidade de produtos, podemos utilizar uma an√°lise de regress√£o ou calcular a elasticidade da convers√£o em rela√ß√£o √† disponibilidade de produtos. Vamos supor que temos uma rela√ß√£o linear entre a disponibilidade de produtos e a convers√£o.

### Passos:

1. **Reunir dados hist√≥ricos de convers√£o e disponibilidade de produtos**.
2. **Calcular a rela√ß√£o entre disponibilidade de produtos e taxa de convers√£o usando uma regress√£o linear**.
3. **Projetar a convers√£o futura com base no aumento da disponibilidade de produtos**.

### Supondo os seguintes dados (exemplo simplificado):

- **Disponibilidade de produtos**: Percentual de disponibilidade de produtos.
- **Taxa de convers√£o**: Percentual de convers√£o de sess√µes em pedidos confirmados (OPC).

Vou usar dados de exemplo simplificados para ilustrar esse processo.

### C√≥digo Python para Proje√ß√£o

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Exemplo de dados hist√≥ricos simplificados
data = {
    'PRODUCT_AVAILABILITY': [85, 87, 88, 89, 90, 92, 94, 95, 96, 98],
    'CONVERSION_RATE': [10, 12, 14, 15, 16, 17, 19, 20, 21, 23]  # Percentual de convers√£o (OPC)
}

df = pd.DataFrame(data)

# Regress√£o linear
X = df['PRODUCT_AVAILABILITY']
y = df['CONVERSION_RATE']
X = sm.add_constant(X)  # Adicionar constante para a regress√£o
model = sm.OLS(y, X).fit()

# Resumo do modelo
print(model.summary())

# Proje√ß√£o de aumento na disponibilidade de produtos
future_availability = np.array([99, 100])  # Exemplos de disponibilidade futura
future_availability = sm.add_constant(future_availability)  # Adicionar constante
predicted_conversion = model.predict(future_availability)

# Resultados projetados
predicted_conversion
```

### An√°lise dos Resultados

O modelo de regress√£o linear fornecer√° uma estimativa da rela√ß√£o entre a disponibilidade de produtos e a taxa de convers√£o. Com base nessa rela√ß√£o, podemos projetar a taxa de convers√£o para diferentes n√≠veis futuros de disponibilidade de produtos.

### Exemplo de Interpreta√ß√£o

Se a regress√£o linear indicar que um aumento de 1% na disponibilidade de produtos resulta em um aumento de 0.5% na taxa de convers√£o, e aumentarmos a disponibilidade de produtos de 90% para 95%, podemos esperar um aumento de 2.5% na taxa de convers√£o.
